```python
import os
import pickle
import pandas as pd
import signalprocessing as sigp
import seaborn as sn
import soundmfccparams as smfccp
```

```python
# Path for RandomForestClassifier
classifierPath = "/home/vaclav/Desktop/Diploma/Notebooks/Classifier/classificator.sav"
# Load the model from disk
loaded_model = pickle.load(open(classifierPath, 'rb'))
```

```python
for filename in os.listdir('/home/vaclav/Desktop/Diploma/Notebooks/Classifier/DataBaseCsv/'):
    signalDataFrame = pd.read_csv('/home/vaclav/Desktop/Diploma/Notebooks/Classifier/DataBaseCsv/'+filename, sep='\t')
    # Predicting the classification
    predictedData = loaded_model.predict(signalDataFrame)
    predictedsignalDataFrame = signalDataFrame.assign(Flag=predictedData)
    lm = sn.lmplot('Total MaxAmp', 'SCO', data=predictedsignalDataFrame, hue="Flag", fit_reg=False, markers=['.','+'])
    lm.fig.suptitle(filename, fontsize=12) 
```
